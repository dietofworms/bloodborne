# Graphing Bloodborne Progress by PSN Trophies
Game progress route: https://bloodborne.wiki.fextralife.com/Game+Progress+Route

```python
import pandas as pd
import seaborn as sns; sns.set()
import matplotlib.pyplot as plt
import matplotlib.ticker as mtick
from adjustText import adjust_text
import numpy as np
from numpy.random import *
%matplotlib inline
```

### Loading the prepared dataset
Data for this section was sourced on 2020-02-03 from: https://psnprofiles.com/trophies/3431-bloodborne.

```python
df = pd.read_csv('dataset.csv')
df.head()
```

### Cleaning and preparing the dataset
Because the dataset was constructed by myself, I have confidence that there are no unexpected errors however some work is needed to get it in the right shape for analysis.

```python
df.dtypes
```

```python
# formatting the game progress column for display in plot
df['progress'] = round(df['progress'] / len(df) * 100, 2)
```

```python
# replacing bool values with text to make the legend more readable
df['dlc'].replace({0 : 'vanilla', 1 : 'dlc'}, inplace=True)
df['optional'].replace({0 : 'mandatory', 1 : 'optional'}, inplace=True)
```

```python
# setting the figure size
plt.rcParams["figure.figsize"] = (26, 22)

# setting font family
plt.rcParams['font.family'] = 'Arial'

# setting plot context
sns.set_context("poster")
```

### Plotting and Exporting the Figure

```python
# create the plot object
ax = sns.scatterplot(x = df['progress'], 
                     y = df['complete'],
                     style = df['dlc'],     # conditional formatting 
                     palette = 'deep',
                     hue = df['optional'],
                     s=400)  # conditional formatting for if the boss is optional

# adding labels to axis and title
ax.set_title('\n How many players have defeated each Bloodborne boss? \n', fontsize = 34, weight='bold')
ax.set_ylabel("\n Percentage of Players Completed \n", fontsize = 26)
ax.set_xlabel("\n Game Progress \n", fontsize = 26)

# formatting the labels to show as %
ax.xaxis.set_major_formatter(mtick.PercentFormatter())
ax.yaxis.set_major_formatter(mtick.PercentFormatter())

# position legend to upper right of the plot and outside the x,y plane
ax.legend(loc='upper right', bbox_to_anchor=(1.14, 1.007), ncol=1)

# add in plot annotation for boss names
for i, txt in enumerate(df['boss']):
    if i == 14: # Mergo's
        ax.annotate(("  " + txt), (df['progress'].iat[i], df['complete'].iat[i]-0.6))
    elif i == 15: # Gehrman
        ax.annotate(("  " + txt), (df['progress'].iat[i], df['complete'].iat[i]-0.8))
    elif i == 16: # Moon Presence
        ax.annotate(("  " + txt), (df['progress'].iat[i], df['complete'].iat[i]+0.2))
    elif i == 17: # Ludwig
        ax.annotate(("  " + txt), (df['progress'].iat[i]-20, df['complete'].iat[i]-0.8))
    elif i == 18: # Laurence
        ax.annotate(("  " + txt), (df['progress'].iat[i]-1.8, df['complete'].iat[i]-1))
    elif i == 19: # Living failures
        ax.annotate(("  " + txt), (df['progress'].iat[i]-12, df['complete'].iat[i]-1))
    elif i == 20: # Lady Maria
        ax.annotate(("  " + (txt[:-24])), (df['progress'].iat[i]-0.6, df['complete'].iat[i]+0.5))
    elif i == 21: # Orphan of kos
        ax.annotate(("  " + txt), (df['progress'].iat[i]-13, df['complete'].iat[i]))
    else: 
        ax.annotate(("  " + txt), (df['progress'].iat[i], df['complete'].iat[i]))


plt.show()
```

```python
# output to file
fig = ax.get_figure()
fig.savefig("plot.png")
```
